```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link="https://www.amazon.in/OnePlus-Silver-Storage-hands-free-capable/product-reviews/B09MQBRCSZ/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

```python
amazon_review=requests.get(link)
```

```python
amazon_review
```

```python
amazon_review.content
```

```python
soup=bs(amazon_review.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
review=soup.find_all('span',{'data-hook':'review-body'})
```

```python
review
```

```python
len(review)
```

```python
customer_review=[]
for i in range(0,len(review)):
    customer_review.append(review[i].get_text())
customer_review
```

```python
customer_review[:]=[review.lstrip('\n') for review in customer_review]
customer_review
```

```python
customer_review[:]=[review.rstrip('\n') for review in customer_review]
customer_review
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
```

```python
df['Review']=customer_review
df
```

```python
df.to_csv(r"C:\Assignment_Excelr\Assignment 11_Text Mining\Review.csv",index=False)
```
